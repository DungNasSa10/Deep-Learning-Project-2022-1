# Crawl pipeline
Command:
```
usage: crawl.py [-h] -f FILEPATH_OR_URL [-d SAVE_DIR] [-sr SAMPLING_RATE] [-rm REMOVE_MP3] [-l TIME_LIMIT]

optional arguments:
  -h, --help            show this help message and exit
  -f FILEPATH_OR_URL, --filepath_or_url FILEPATH_OR_URL
                        Can be file path or google drive shared url
  -d SAVE_DIR, --save_dir SAVE_DIR
                        Directory to save .wav files
  -sr SAMPLING_RATE, --sampling_rate SAMPLING_RATE
                        Sampling rate of output .wav files
  -rm REMOVE_MP3, --remove_mp3 REMOVE_MP3
                        Remove downloaded mp3
  -l TIME_LIMIT, --time_limit TIME_LIMIT
                        time limit in seconds
```

Example:
```
!python src/crawl.py \
```
